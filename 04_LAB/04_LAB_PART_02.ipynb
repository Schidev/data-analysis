```python
from transformers import pipeline
```

```python
summarizer = pipeline("summarization", model='ukr-models/uk-summarizer')
```

```python
text = """Видозмінений вуглець - книга, присвячена захопливому дослідженню в галузі матеріалів майбутнього. Автор, експерт у галузі нанотехнологій, представляє читачам захопливе занурення у світ видозміненого вуглецю та його потенціалу для трансформації технологій.
Книга починається з основ, розкриваючи унікальні властивості видозміненого вуглецю, його структуру та вплив на фізичні й хімічні властивості матеріалу. Піднімаючись від атомарного рівня до макроскопічних явищ, автор переконливо показує, як ці властивості можуть бути використані для революціонізації різних галузей."""
```

```python
summary = summarizer(text, max_length=150, min_length=45, length_penalty=2.0, num_beams=4, early_stopping=True)

print(summary[0]['summary_text'])
```
